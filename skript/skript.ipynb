```python
def pica_schreiben(row):
    pica = f"""0500 Da
0501 Text$btxt
0502 ohne Hilfsmittel zu benutzen$bn
0503 Blatt$bnb
0600 yy
1100 {row.jahr}
"""
    if pd.notna(row.ezth):
        pica +=f"1110 {row.ezth}$4ezth\n"

        pica +=f"""1130 TB-papier
1131 {row.form}
1132 a1-analog;a2-hand;f1-text;e2-uv;o-org;v-ganz
1500 {row.sprache}
1700 {row.land}
"""

    if pd.notna(row.verf):
        pica +=f"3000 {row.verf}$BVerfasser$4aut\n"
    if pd.notna(row.urheber) and pd.notna(row.urheber2):
        pica +=f"3010 {row.urheber}$BAdressat$4rcp\n3010 {row.urheber2}$BAdressat$4rcp\n"
    elif pd.notna(row.urheber) and pd.isna(row.urheber2):
        pica +=f"3010 {row.urheber}$BAdressat$4rcp\n"


    if pd.notna(row.urheber) and pd.notna(row.urheber2):
        pica +=f"3100 {row.urheber}$BVerfasser$4aut\n3010 {row.urheber2}$BMitwirkender$4ctb\n" 
    elif pd.notna(row.urheber) and pd.isna(row.urheber2):
        pica +=f"3100 {row.urheber}$BVerfasser$4aut\n"
    
    pica += f"4000 {row.titel}\n"

    if pd.notna(row.ort):
        pica +=f"4046 {row.ort}\n"
    
        pica +=f"""4060 {row.umfang}
4062 {row.blatt}
"""

    if pd.notna(row.beil):
        pica +=f"4063 {row.beil}\n"

    pica +=f"4105 !!\n"

    if pd.notna(row.bemerkung):
        pica +=f"4201 {row.bemerkung}\n"

    pica +=f"5590 [Objektgattung]\n"
    gattungsliste = row.gattung1.split('!')
    counter = 0
    for idn in gattungsliste:
        if counter % 2 != 0:
            pica +=f"5590 !{idn}!\n"
        counter += 1

        
    pica += "\t\n"
    return pica
```

```python
for index, row in df.sample(10).iterrows():
    print(pica_schreiben(row))
```
